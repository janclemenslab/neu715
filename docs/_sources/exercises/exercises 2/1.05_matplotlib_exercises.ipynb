## Exercise 2: Overlaying multiple curves
We have recorded from another cell and want to compare the intensity tuning curves for both neurons.

2. the color for the curve of the first cell is red, and that the color of the second cell is green.
3. there is a legend indicating which is cell is which (call them "Cell 1" and "Cell 2")

## Exercise 3: Subplots
Now, instead of overlaying the curves, we would like to plot each in a separate subplot.

Generate a figure with two subplots. Set the limits of the x and y axis for each subplot so that they are the same for both plots.

```python
firing_rate1_Hz = [0, .06, 0.5, 1.1, 1.25, 1.3]
firing_rate2_Hz = [0, 0, 0.1, 0.5, 0.9, 0.89]
sound_intensity_dB = [40, 50, 60, 70 , 80, 90]

# your solution
plt.figure(figsize=(12, 3))
plt.subplot(2, 2, 1)
plt.plot(sound_intensity_dB, firing_rate1_Hz, 'o-k')
plt.xlabel('SI [dB]')
plt.ylabel('FR [Hz]')

plt.subplot(2, 2, 4)
plt.plot(sound_intensity_dB, firing_rate2_Hz, 'o-g')
plt.xlabel('SI [dB]')
plt.ylabel('FR [Hz]')
plt.legend(['Cell 1', 'Cell 2'])
```
