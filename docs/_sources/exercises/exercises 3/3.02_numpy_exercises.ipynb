## Exercise 3: Z-scoring our data
In a previous exercised, we have z-scored some temperature data using core python and a bunch of for loops.

Remember that z-scoring amounts to for each value: subtracting the mean, $M$, and dividing by the standard deviation, $S$:
$z_i = (x_i - M) / S$

```python
# EXECUTE BUT DO NOT CHANGE THIS CELL
temperature_anomalies = [-0.19,-0.177,-0.308,-0.289,-0.193,-0.057,0.077,0.026,0.029,0.068,0.256, 0.342, 0.591, 0.792]
print(temperature_anomalies)

# compute mean
mean = 0
for ta in temperature_anomalies:
    mean = mean + ta
mean = mean / len(temperature_anomalies)

# compute std
std = 0
for ta in temperature_anomalies:
    std = std + (ta - mean) ** 2
std = (std / len(temperature_anomalies)) ** (1/2)

# z-score - subtract the mean and divide by the standard deviation
ta_zscored = []
for ta in temperature_anomalies:
    ta_zscored.append( (ta - mean) / std)

ta_zscored
```

This is quite a bit of code. Use Numpy to get z-score the data in 1-2 lines of code, without any for loops. You can use `np.mean` and `np.std` to compute the mean and standard deviation

```python
# EXECUTE BUT DO NOT CHANGE THIS CELL
temperature_anomalies = [-0.19,-0.177,-0.308,-0.289,-0.193,-0.057,0.077,0.026,0.029,0.068,0.256, 0.342, 0.591, 0.792]
print(temperature_anomalies)

# your solution here
import numpy as np

ta_zscored = (temperature_anomalies - np.mean(temperature_anomalies)) / np.std(temperature_anomalies)
ta_zscored
```
