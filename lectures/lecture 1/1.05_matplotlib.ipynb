Plotting is not part of core python - we have to use an external library for that: [Matplotlib](https://matplotlib.org).


Matplotlib comes installed with the anaconda distribution and the pyplot module is the most user friendly.
To make it available in the notebook, we need to "import" it.

What you need to know for now (More details on how to work with external libraries next week): 

```import matplotlib.pyplot as plt``` 

imports the pyplot module of matplotlib and makes everything under it available via the `plt` shortcut:

````python
import matplotlib.pyplot as plt

plt.plot(voltage_trace)  # plot the voltage traces (list of voltage values)
plt.show()  # show the plot here
````

By default, pyplot produces reasonably useful plots - but we can tweak and annotate them to make them even better:

### Size and annotate - figsize, xlabel, ylabel, title

````python
plt.figure(figsize=(20, 2))  # create a figure with a 1:10 aspect ratio
plt.plot(voltage_trace, 'k')  # plot the voltage trace as a black line
plt.ylabel('Membrane voltage [mV]')  # label the y axis
plt.xlabel('Time [ms]')  # label the y axis
plt.title(' Recording')  # plot title
plt.show()
````

### Specifiy the values on the x axis

````python
plt.figure(figsize=(20, 2))
plt.plot(times_seconds, voltage_trace, 'k')  # use seconds as x-values
plt.ylabel('Membrane voltage [mV]')
plt.xlabel('Time [seconds]')  # change the time units
plt.title(' Recording')
plt.show()
````

### Change axis limits - xlim, ylim

````python
plt.figure(figsize=(20, 2))
plt.plot(times_seconds, voltage_trace, 'k')
plt.ylabel('Membrane voltage [mV]')
plt.xlabel('Time [s]')
plt.title('Subthreshold voltages')
plt.ylim(-80, -20)  # set limits of y axis
plt.xlim(0)  # set only lower limit of x axis
plt.show()
````

### Change/combine markers

````python
plt.figure(figsize=(20, 2))
plt.plot(times_seconds, voltage_trace, 'o-', c='k')  # plot voltage traces as lines and dots
plt.ylabel('Membrane voltage [mV]')
plt.xlabel('Time [s]')
plt.title('Recording')
plt.show()
````

### Compose/overlay multiple plots

````python
plt.figure(figsize=(20, 2))
plt.plot(times_seconds, voltage_trace, '-', c='k')  # voltage trace as line in black
plt.plot(times_seconds, voltage_trace, 'o', c='r', alpha=0.2)  # individual voltage values as dots, `alpha` controls opacity
plt.ylabel('Membrane voltage [mV]')  # label the y axis
plt.xlabel('Time [s]')  # label the y axis
plt.title('Recording')
plt.show()
````

### Multi-panel plots, tick spacing - subplot, yticks

````python
plt.figure(figsize=(20, 4))
plt.subplot(2, 1, 1)  # first panel
plt.plot(times_seconds, voltage_trace, 'k')
plt.xlim(0)
plt.xticks([])  # remove xticks from the upper panel
plt.title('Recording')

plt.subplot(2,1,2)  # second panel
plt.plot(times_seconds, voltage_trace, 'k')
plt.ylabel('Membrane voltage [mV]')
plt.xlabel('Time [s]')
plt.title('Subthreshold voltages')
plt.ylim(-90, -30)
plt.xlim(0)
plt.yticks([-80, -60, -40, -20])  # change tick spacing in the lower panel
plt.show()
````

### Legends

````python
plt.figure(figsize=(20, 2))
plt.plot(times_seconds, voltage_trace, '-', c='k', label='line')  # define legend label
plt.plot(times_seconds, voltage_trace, 'o', c='r', alpha=0.2, label='dots')  # define legend label
plt.ylabel('Membrane voltage [mV]')  # label the y axis
plt.xlabel('Time [s]')  # label the y axis
plt.title('Recording')
plt.legend()  # add the legend - alternatively: plt.legend(['line', 'dots'])
plt.show()
````
